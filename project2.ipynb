```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = Image.open("image1.png")
image_array = np.array(image)

print("Image dimensions:", image_array.shape)
image
```

```python
image_shape = image_array.shape
print("Image dimensions:", image_shape)
```

```python
is_integer = np.issubdtype(image_array.dtype, np.integer)
print("Is the image integer type?", is_integer)
```

```python
gray_image = np.dot(image_array[..., :3], [0.2989, 0.587, 0.114]).astype(np.uint8)

plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
lightened = np.clip(gray_image + 50, 0, 255).astype(np.uint8)

plt.imshow(lightened, cmap='gray')
plt.title("Lightened Image")
plt.axis("off")
plt.show()
```

```python
darkened = np.clip(gray_image - 50, 0, 255).astype(np.uint8)

plt.imshow(darkened, cmap='gray')
plt.title("Darkened Image")
plt.axis("off")
plt.show()
```

```python
contrast_adjusted = np.clip(1.5 * (gray_image - 128) + 128, 0, 255).astype(np.uint8)

plt.imshow(contrast_adjusted, cmap='gray')
plt.title("Contrast Adjusted Image")
plt.axis("off")
plt.show()
```

```python
cropped_image = gray_image[50:200, 50:200] 

plt.imshow(cropped_image, cmap='gray')
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
flipped_horizontally = np.fliplr(gray_image)

plt.imshow(flipped_horizontally, cmap='gray')
plt.title("Flipped Horizontally")
plt.axis("off")
plt.show()
```

```python
rotated_90 = np.rot90(gray_image)

plt.imshow(rotated_90, cmap='gray')
plt.title("Rotated 90°")
plt.axis("off")
plt.show()
```

```python
# 9. Display All Modified Images
fig, axes = plt.subplots(2, 4, figsize=(12, 6))

axes[0, 0].imshow(gray_image, cmap='gray')
axes[0, 0].set_title("Grayscale")

axes[0, 1].imshow(lightened, cmap='gray')
axes[0, 1].set_title("Lightened")

axes[0, 2].imshow(darkened, cmap='gray')
axes[0, 2].set_title("Darkened")

axes[0, 3].imshow(contrast_adjusted, cmap='gray')
axes[0, 3].set_title("Contrast Adjusted")

axes[1, 0].imshow(cropped_image, cmap='gray')
axes[1, 0].set_title("Cropped")

axes[1, 1].imshow(flipped_horizontally, cmap='gray')
axes[1, 1].set_title("Flipped Horizontally")

axes[1, 2].imshow(rotated_90, cmap='gray')
axes[1, 2].set_title("Rotated 90°")

for ax in axes.flat:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
